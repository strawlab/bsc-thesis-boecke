# This creates the file with all the clippings from the original files on the basis of my summary file:

```python
def uploader_sorter_creator (row): 
    
    
    #print(row)    
    fly_at = row['fly_number']
    extraction_fname = "/Users/ambra-stella/bachelor/DATA/processed/smart_load{}.csv".format(fly_at)
    #print(os.path.isfile(extraction_fname))
    try:
        extractee = pd.read_csv(extraction_fname, usecols = ['frame', 'timestamp', 'x', 'y', 'xvel', 'yvel', 't', 'x_px', 'y_px', 'orientation_radians_mod_pi', 'contrast','luminance', 'center_x', 'center_y', 'stage', 'direction', 'ON','direction_change', 'area_change', 'area', 'stopped', 'ntimes'], low_memory=False)
        #print(extractee.shape)
    except ValueError:
        return None
        
    #print (extractee[extractee.t == row['t']].t)
    tstart_whole = extractee[extractee.t == row['t']].t - 4 
    tstart_whole = tstart_whole.iloc[-1]
    #print("start", tstart_whole)
    tstopp_whole = extractee[extractee.t == row['t']].t + 6
    tstopp_whole = tstopp_whole.iloc[-1]
    #print("stop ", tstopp_whole)
    
    #for t0,t1 in zip(tstart_whole, tstopp_whole):
    extract = extractee[(extractee.t>tstart_whole)&(extractee.t<tstopp_whole)].copy()
    
    
    extract['distraction_index'] = row['distraction_index']
    extract['fly_number'] = fly_at
    extract['timing'] = ""
    
    
    tstart_before = tstart_whole 
    tstopp_before = tstart_whole + 4

    extract.loc[(extract.t>tstart_before)&(extract.t<tstopp_before), "timing"] = 'before'

    tstart_bar = tstart_whole + 4 
    tstopp_bar = tstopp_whole -4
    #for t0,t1 in zip(tstart_bar, tstopp_bar):
    extract.loc[(extract.t>tstart_bar)&(extract.t<tstopp_bar), "timing"] = 'bar'
    
    tstart_after = tstopp_whole - 4  
    tstopp_after = tstopp_whole
    #for t0,t1 in zip(tstart_after, tstopp_after):
    extract.loc[(extract.t>tstart_after)&(extract.t<tstopp_after), "timing"] = 'after'
    
    return extract



#test= distr_summary.apply(uploader_sorter_creator, axis =1)
```

# This goes through all the original files and saves the clippings in Path: (CHECK FOR RIGHT PATH!!)

```python
for i, row in distr_summary.iterrows():
    current_result = uploader_sorter_creator(row)
    print(type(current_result))
    try:
        current_result.to_csv("/Users/ambra-stella/bachelor/analysis/TNTe/clipped_reaction.csv",  mode = 'a', header=False, index=False)
    except AttributeError:
        pass
```
